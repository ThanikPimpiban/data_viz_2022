Quiz กลุ่ม final 

    * เขียน function ที่รับ input เป็น output ของ boxplot แล้วแสดงค่า min, max ,q1, q2 , q3

    * วาด box plot เปรียบเทียบ การกระจายตัวของข้อมูล PetalLength ของดอก iris ทั้ง 3 ชนิด

```python
ybp = plt.boxplot(df['SepalLength'],vert=False);
```

## 1

```python
## หาค่า
min = ybp['caps'][0].get_xdata()[0]
max = ybp['caps'][1].get_xdata()[0]
q1 = ybp['boxes'][0].get_xdata()[0]
q2 = ybp['medians'][0].get_xdata()[0]
q3 = ybp['boxes'][0].get_xdata()[2]


## create function
def function():
    print(f'min = {min}')
    print(f'max = {max}')
    print(f'q1 = {q1}')
    print(f'q2 = {q2}')
    print(f'q3 = {q3}')
```
