# ระบุสีให้แต่ละจุด

```python
set(df['Name'])
```

```python
df2=df.replace({'Iris-setosa':'r', 'Iris-versicolor':'g','Iris-virginica':'b'}) ##แทนชื่อดอกไม้ ด้วย สี 
```

```python
df2.head()
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
```

# เพิ่มรายละเอียดกราฟ

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
```

# ชื่อ marker

```python
plt.scatter(df['PetalLength'][0:50],df['SepalLength'][0:50],c=df2['Name'][0:50],label='Iris-setosa')
plt.scatter(df['PetalLength'][50:100],df['SepalLength'][50:100],c=df2['Name'][50:100],label='Iris-versicolor')
plt.scatter(df['PetalLength'][100:150],df['SepalLength'][100:150],c=df2['Name'][100:150],label='Iris-virginica')
plt.legend()
plt.title('Scatter of Iris-Data')
plt.xlabel('PetalLength')
plt.ylabel('SepalLength');
```

# ขนาดของ marker ใช้ค่าใน column

```python
plt.scatter(df['PetalLength'][0:50],df['SepalLength'][0:50],c=df2['Name'][0:50],label='Iris-setosa',s=df2['PetalWidth'][0:50])
plt.scatter(df['PetalLength'][50:100],df['SepalLength'][50:100],c=df2['Name'][50:100],label='Iris-versicolor',s=df2['PetalWidth'][50:100])
plt.scatter(df['PetalLength'][100:150],df['SepalLength'][100:150],c=df2['Name'][100:150],label='Iris-virginica',s=df2['PetalWidth'][100:150])
plt.legend()
plt.title('Scatter of Iris-Data')
plt.xlabel('PetalLength')
plt.ylabel('SepalLength');
```

```python
plt.scatter(df['PetalLength'][0:50],df['SepalLength'][0:50],c=df2['Name'][0:50],label='Iris-setosa',s=20*df2['PetalWidth'][0:50])
plt.scatter(df['PetalLength'][50:100],df['SepalLength'][50:100],c=df2['Name'][50:100],label='Iris-versicolor',s=20*df2['PetalWidth'][50:100])
plt.scatter(df['PetalLength'][100:150],df['SepalLength'][100:150],c=df2['Name'][100:150],label='Iris-virginica',s=20*df2['PetalWidth'][100:150])
plt.legend()
plt.title('Scatter of Iris-Data')
plt.xlabel('PetalLength')
plt.ylabel('SepalLength');
```

# ปรับความโปร่งใสของ marker

```python
plt.scatter(df['PetalLength'][0:50],df['SepalLength'][0:50],alpha=0.5,c=df2['Name'][0:50],label='Iris-setosa',s=20*df2['PetalWidth'][0:50])
plt.scatter(df['PetalLength'][50:100],df['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='Iris-versicolor',s=20*df2['PetalWidth'][50:100])
plt.scatter(df['PetalLength'][100:150],df['SepalLength'][100:150],alpha=0.5,c=df2['Name'][100:150],label='Iris-virginica',s=20*df2['PetalWidth'][100:150])
plt.legend()
plt.title('Scatter of Iris-Data')
plt.xlabel('PetalLength')
plt.ylabel('SepalLength');
```
